```python
df = pd.read_csv('/kaggle/input/diabetes-prediction-dataset/diabetes_prediction_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Define the mapping
mapping = {'Male': 0, 'Female': 1}

# Fill NaN values in 'gender' column with 'unknown'
df['gender'] = df['gender'].fillna('unknown')

# Apply label encoding
df['gender_encoded'] = df['gender'].map(mapping)
```

```python
df
```

```python
df.isnull().sum()
```

```python
for column in df.columns[df.isnull().any()]:
    mode_value = df[column].mode()[0]
    df[column]=df[column].fillna(mode_value)
```

```python
df.drop('gender' , axis = 1)
```

```python
df['smoking_history'].unique()
```

```python

# Extract the smoking_history column
smoking_history_column = df['smoking_history'].values.reshape(-1, 1)

# Define the ordinal encoder
ordinal_encoder = OrdinalEncoder(categories=[['never', 'No Info', 'not current', 'former', 'current', 'ever']])

# Fit and transform the encoder to convert smoking history values
encoded_smoking_history = ordinal_encoder.fit_transform(smoking_history_column)

# Update the DataFrame with the encoded values
df['smoking_history_encoded'] = encoded_smoking_history
```

```python
df = df.drop('gender', axis = 1)
```

```python
 df = df.drop('smoking_history', axis = 1)
```

```python
df
```

```python

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

# Define which columns to scale (all numerical columns)
numerical_columns = ['age', 'hypertension', 'heart_disease', 'bmi', 'HbA1c_level', 
                     'blood_glucose_level', 'diabetes', 'gender_encoded', 'smoking_history_encoded']

# Scale the numerical columns
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
df
```

```python
df.isnull().sum()
```

```python
X = df[['age' , 'hypertension',	'heart_disease','bmi','HbA1c_level','blood_glucose_level','gender_encoded',	'smoking_history_encoded']]
```

```python
Y = df['diabetes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split( X ,Y , test_size = 0.3 , random_state = 42)
```

```python
X_train.shape , X_test.shape , Y_train.shape , Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
lrmodel = LogisticRegression()
```

```python
lrmodel.fit(X_train , Y_train)
```

```python
y_pred = lrmodel.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)
conf_matrix = confusion_matrix(Y_test, y_pred)
classification_rep = classification_report(Y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```
